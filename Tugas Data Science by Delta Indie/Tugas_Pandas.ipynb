```python
# Import library pandas
import pandas as pd
```

```python
# Import dataset
df=pd.read_csv("data_penjualan.csv",delimiter=";")
df
```

```python
# Memfilter dataset untuk menampilkan produk 'FoodpakMatte245'
fm=df[df["Jenis Produk"]=="FoodpakMatte245"].copy()
fm
```

```python
# Mengetahui informasi umum untuk data jenis produk 'FoodpakMatte245'
fm.info()
```

Terdapat 54 transaksi produk FoodpakMatte245.

1. Hitung rata-rata penjualan dari produk FoodpakMatte245!

```python
# Rata-rata penjualan produk FoodpakMatte245
round(fm["Total"].mean(),2)
```

Rata-rata penjualan produk FoodpakMatte245 yaitu sebesar Rp 2.861.111,11.

2. Di tanggal berapakah transaksi termahal terjadi?

```python
# Menampilkan hanya satu tanggal pertama terjadinya transaksi termahal
fm.loc[fm["Total"].idxmax(),"Tanggal"]
```

```python
# Cara lain untuk menampilkan tanggal transaksi termahal

# Mendefinisikan nilai maksimum
maks_total=fm["Total"].max()
maks_total
```

```python
# Menampilkan seluruh tanggal dengan transaksi termahal
fm[fm["Total"]==maks_total]
```

Terdapat tiga transaksi produk FoodpakMatte245 dengan total transaksi termahal sebesar Rp 7.500.000,00, yaitu pada tanggal 23 Maret 2023, 25 April 2023, dan 6 Juni 2023.

3. Berapakah jumlah orderan terbanyak di produk FoodpakMatte245?

```python
# Jumlah order terbanyak produk FoodpakMatte245
fm["Jumlah Order"].max()
```

Jumlah order terbanyak untuk produk FoodpakMatte245 yaitu sebanyak 6000 unit.

4. Berapa nilai terendah dari total pembelian yang terjadi di bulan 10 tahun 2023?

```python
# Mengubah kolom 'Tanggal' menjadi format datetime
fm["Tanggal"]=pd.to_datetime(fm["Tanggal"],format="%d/%m/%Y")
fm["Tanggal"].info()
```

```python
# Memfilter transaksi di bulan 10 tahun 2023
fm1023=fm[(fm["Tanggal"].dt.month==10) & (fm["Tanggal"].dt.year==2023)]
fm1023
```

```python
# Menampilkan transaksi terendah di bulan 10 tahun 2023
fm1023["Total"].min()
```

Nilai terendah dari total pembelian yang terjadi di bulan 10 tahun 2023 yaitu sebesar Rp 650.000,00.

5. Di tanggal berapa nilai harga FoodpakMatte245 tertinggi?

```python
# Menampilkan hanya satu tanggal pertama terjadinya harga tertinggi
fm.loc[fm["Harga"].idxmax(),"Tanggal"]
```

```python
# Cara lain untuk menampilkan tanggal dengan harga tertinggi

# Mendefinisikan nilai maksimum harga
maks_harga=fm["Harga"].max()
maks_harga
```

```python
# Menampilkan tanggal dengan harga tertinggi
fm[fm["Harga"]==maks_harga]
```

Transaksi FoodpakMatte245 dengan harga tertinggi sebesar Rp 2.400,00 terjadi pada tanggal 20 Oktober 2023.
